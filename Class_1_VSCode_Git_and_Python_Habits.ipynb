## 1. Why use VS Code?

VS Code is useful because it gives students one place for:
- editing code,
- running Python,
- using the terminal,
- viewing folders and files,
- working with notebooks,
- managing Git.

In this course, students may use both:
- **Python scripts**: files ending in `.py`
- **Jupyter notebooks**: files ending in `.ipynb`

A good habit is to think of VS Code as the main workspace, not just a text editor.



## 2. Basic VS Code workflow

A typical workflow looks like this:

1. Create a folder for the course or project
2. Open that folder in VS Code
3. Create a Python file or notebook
4. Open the terminal inside VS Code
5. Run the code
6. Save often
7. Use Git to keep track of changes

### Important VS Code ideas
- **Explorer:** shows files and folders
- **Editor:** where you write code
- **Terminal:** where you run commands
- **Extensions:** add features like Python, Jupyter, GitHub support
- **Source Control tab:** lets you see Git changes

## 4. Virtual environments

A **virtual environment** keeps project packages separate from the rest of the computer.

This is a very good habit because different projects may need different package versions.

### Example commands

On macOS/Linux:
```bash
python3 -m venv .venv
source .venv/bin/activate
pip install numpy matplotlib jupyter
```

On Windows:
```bash
python -m venv .venv
.venv\Scripts\activate
pip install numpy matplotlib jupyter
```

In VS Code, after creating the environment, select the correct Python interpreter.



## 5. Git as a programming habit, not an afterthought

Git is a version-control system. It helps you:
- save meaningful checkpoints,
- recover old versions,
- see what changed,
- collaborate later,
- avoid losing work.

Students often wait too long to learn Git. It is better to start with a few simple habits immediately.



## 6. Minimal Git workflow

Inside the project folder, a simple start is:

```bash
git init
git add .
git commit -m "Initial commit"
```

Then, after a useful change:

```bash
git add .
git commit -m "Added first motion plot"
```

### Good Git habits
- commit after a meaningful change,
- write short but descriptive commit messages,
- do not wait until code becomes messy,
- keep code and notes in the same project folder when appropriate.



- **Git records project history**
- **A commit is a saved checkpoint**
- **Small frequent commits are better than rare giant commits**


## 8. Python basics: variables, comments, and units

Even in VS Code, the scientific habits are the same:
- use clear variable names,
- write comments,
- keep track of units,
- test code in small pieces.


````python

mass = 2.0          # kg
gravity = 9.81      # m/s^2
weight = mass * gravity

print("Mass =", mass, "kg")
print("Weight =", weight, "N")
````


### Good habit 1: descriptive names
Prefer:
- `mass`
- `gravity`
- `time_s`
- `position_m`

instead of vague names like:
- `a`
- `tmp`
- `val1`



### Good habit 2: keep units visible
Code becomes much easier to trust when units are clear.


````python

distance_m = 1500
time_s = 180
speed_m_per_s = distance_m / time_s

print("Speed =", speed_m_per_s, "m/s")
````


## 9. Arrays with NumPy

In physics, we usually want many values at once.


````python

t = np.linspace(0, 10, 6)
x = 3 * t + 2

print("t =", t)
print("x =", x)
````


## 10. Plotting in Python

Plotting is one of the main reasons to learn Python in physics.


````python

x0 = 2.0
v = 3.0
t = np.linspace(0, 10, 200)
x = x0 + v * t

plt.figure(figsize=(7, 4))
plt.plot(t, x, linewidth=2)
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.title("Motion with Constant Velocity")
plt.grid(True)
plt.show()
````


### Good habit 3: clean plots
Every scientific plot should usually include:
- axis labels,
- units,
- title,
- readable formatting.



## 11. Small physics example: simple harmonic motion

A simple harmonic motion can be written as

$$
x(t) = A\cos(\omega t).
$$


````python

A = 0.5
omega = 2 * np.pi
t = np.linspace(0, 2, 500)
x = A * np.cos(omega * t)

plt.figure(figsize=(7, 4))
plt.plot(t, x, linewidth=2)
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.title("Simple Harmonic Motion")
plt.grid(True)
plt.show()
````
