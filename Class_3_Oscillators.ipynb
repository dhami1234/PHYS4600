## 1. Rewriting a second-order ODE

A second-order ODE such as

$$
\ddot{x} = a(t,x,\dot{x})
$$

can be rewritten by defining

$$
v=\dot{x}.
$$

Then we solve the first-order system

$$
\dot{x}=v,\qquad \dot{v}=a(t,x,v).
$$


## 2. Euler and midpoint for second-order equations

For

$$
\dot{x}=v,\qquad \dot{v}=a(t,x,v),
$$

Euler gives

$$
x_{n+1}=x_n+h v_n,\qquad
v_{n+1}=v_n+h\,a(t_n,x_n,v_n).
$$

Midpoint gives

$$
x_{n+\frac12}=x_n+\frac{h}{2}v_n,\qquad
v_{n+\frac12}=v_n+\frac{h}{2}a(t_n,x_n,v_n),
$$

then

$$
x_{n+1}=x_n+h\,v_{n+\frac12},\qquad
v_{n+1}=v_n+h\,a\!\left(t_n+\frac{h}{2},x_{n+\frac12},v_{n+\frac12}\right).
$$


```python
def euler_second_order(acceleration, t0, tf, x0, v0, h):
    t = np.arange(t0, tf+h, h)
    x = np.zeros_like(t)
    v = np.zeros_like(t)
    x[0] = x0
    v[0] = v0
    for n in range(len(t) - 1):
        x[n+1] = x[n] + h*v[n]
        v[n+1] = v[n] + h*acceleration(t[n],x[n],v[n]) 
    
    return t, x, v

def midpoint_second_order(acceleration, t0, tf, x0, v0, h):
    t = np.arange(t0, tf+h, h)
    x = np.zeros_like(t)
    v = np.zeros_like(t)
    x[0] = x0
    v[0] = v0
    for n in range(len(t) - 1):
        x_mid = x[n] + 0.5*h*v[n]
        v_mid = v[n] + 0.5*h*acceleration(t[n], x[n], v[n])

        x[n+1] = x[n] + h * v_mid
        v[n+1] = v[n] + h * acceleration(t[n] + 0.5*h , x_mid, v_mid)
    
    return t, x, v

```

## 3. Ideal harmonic oscillator

The ideal oscillator satisfies

$$
m\ddot{x}+kx=0
$$

or

$$
\ddot{x}=-\omega_0^2 x,\qquad \omega_0=\sqrt{\frac{k}{m}}.
$$


```python
m = 1.0
k = 1.0
omega0 = np.sqrt(k / m)

t0 = 0.0
tf = 20.0
x0 = 1.0
v0 = 0.0
h = 0.05

def a_ideal(t, x, v):
    return -omega0*omega0*x

t_e, x_e, v_e = euler_second_order(a_ideal, t0, tf, x0, v0, h)
t_m, x_m, v_m = midpoint_second_order(a_ideal, t0, tf, x0, v0, h)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(t_e, x_e, label="Euler")
plt.plot(t_m, x_m, label="Midpoint")
plt.xlabel("Time")
plt.ylabel("Position")
plt.title("Ideal Harmonic Oscillator: Position vs Time")
plt.grid(True)
plt.legend()
plt.show()
```

```python
E_e = 0.5*m*v_e*v_e + 0.5*k*x_e*x_e
E_m = 0.5*m*v_m*v_m + 0.5*k*x_m*x_m

plt.figure(figsize=(8, 5))
plt.plot(t_e, E_e, label="Euler")
plt.plot(t_m, E_m, label="Midpoint")
plt.xlabel("Time")
plt.ylabel("Energy")
plt.title("Energy in the Ideal Harmonic Oscillator")
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.plot(x_e, v_e, label="Euler")
plt.plot(x_m, v_m, label="Midpoint")
plt.xlabel("x")
plt.ylabel("v")
plt.title("Phase Space: Ideal Harmonic Oscillator")
plt.grid(True)
plt.legend()
plt.axis("equal")
plt.show()
```

```python
def animate_1d_motion(x, t, title="1D Motion", trace=False, interval=30):
    fig, ax = plt.subplots(figsize=(8, 2.5))
    xmin = 1.2 * np.min(x)
    xmax = 1.2 * np.max(x)
    if xmin == xmax:
        xmin -= 1
        xmax += 1
    ax.set_xlim(xmin, xmax)
    ax.set_ylim(-1, 1)
    ax.set_yticks([])
    ax.set_xlabel("x")
    ax.set_title(title)
    ax.grid(True)
    ax.plot([xmin, xmax], [0, 0], "k--", alpha=0.4)
    point, = ax.plot([], [], "o", markersize=10)
    trail, = ax.plot([], [], "-", alpha=0.7)

    def init():
        point.set_data([], [])
        trail.set_data([], [])
        return point, trail

    def update(frame):
        point.set_data([x[frame]], [0])
        if trace:
            trail.set_data(x[:frame+1], np.zeros(frame+1))
        return point, trail

    ani = FuncAnimation(fig, update, frames=len(t), init_func=init, interval=interval, blit=True)
    plt.close(fig)
    return ani
```

```python
ani_ideal = animate_1d_motion(x_m, t_m, title="Ideal Harmonic Oscillator", trace=True)
HTML(ani_ideal.to_jshtml())
```

## 4. Damped harmonic oscillator

The damped oscillator satisfies

$$
m\ddot{x}+c\dot{x}+kx=0.
$$

Writing

$$
\ddot{x}+2\gamma \dot{x}+\omega_0^2 x=0,\qquad \gamma=\frac{c}{2m},
$$

we classify the motion as:
- underdamped if $\gamma<\omega_0$
- critically damped if $\gamma=\omega_0$ ($c = 2 * \sqrt(k * m)$)
- overdamped if $\gamma>\omega_0$


```python

cs = ["three values"]
plt.figure(figsize=(8, 5))
for c in cs:
    def a_damped(t, x, v, c=c):
        return
    t_d, x_d, v_d = midpoint_second_order(a_damped, 0.0, 20.0, x0, v0, h)
    plt.plot(t_d, x_d, label=f"c = {c:.2f}")

plt.xlabel("Time")
plt.ylabel("Position")
plt.title(f"Damped Harmonic Oscillator: c = {c:.2f}")
plt.grid(True)
plt.legend()
plt.show()
```

```python
ani_under = animate_1d_motion(x_d, t_d, title="Underdamped Oscillator", trace=True)
HTML(ani_under.to_jshtml())
```

## 7. Amplitude vs driving frequency

```python
omega_values = np.linspace(0.2, 1.8, 80)
amplitudes = []

t_start = 0.0
t_end = 480.0
h_amp = 0.02

for omega_val in omega_values:
    def a_temp(t, x, v, omega_val=omega_val):
        return 
    t_tmp, x_tmp, v_tmp = midpoint_second_order(a_temp, t_start, t_end, x0, v0, h_amp)
    keep = t_tmp > 0.7 * t_end
    amp = 0.5 * (np.max(x_tmp[keep]) - np.min(x_tmp[keep]))
    amplitudes.append(amp)

plt.figure(figsize=(8, 5))
plt.plot(omega_values, amplitudes)
plt.axvline(np.sqrt(k / m), linestyle="--", label=r"$\omega_0$")
plt.xlabel(r"Driving frequency $\omega$")
plt.ylabel("Steady-state amplitude")
plt.title("Amplitude vs Driving Frequency")
plt.grid(True)
plt.legend()
plt.show()

imax = np.argmax(amplitudes)
print(f"Maximum amplitude occurs near Omega = {omega_values[imax]:.3f}")
```
